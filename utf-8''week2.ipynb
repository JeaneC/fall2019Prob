### Bernoulli Distribution
The Bernoulli random variable records the result of a single trial with two possible outcomes. 

Let 0 $\leq$ p $\leq$ 1. A random variable X has the Bernoulli distribution with success probability p if X is {0, 1}-valued and satisfies P(X = 1) = p and P(X = 0) = 1 âˆ’ p. 

In the example below

- X = Heads or Tails
- An outcome of of 0 = Tails
- An outcome of 1 = Heads
- P(Heads) = 0.65
- P(Tails) = 1 - 0.65 = 0.35

```r
# Note that the graphs below reflect the probability of getting heads or tails.

p <- 0.65

a <- rbinom(100, 1, p)
hist(a, main="100 Coin Flips", xlab = "0 = Tails, 1 = Heads", ylim = c(0, 100))
# Expecting frequency of 0 = Tails to be 35, Heads to be 65

b <- rbinom(150, 1, p)
hist(b, main="150 Coin Flips", xlab = "0 = Tails, 1 = Heads", ylim = c(0, 150))

c <- rbinom(200, 1, p)
hist(c, main="200 Coin Flips", xlab = "0 = Tails, 1 = Heads", ylim = c(0, 200))

```

### Binomial Distribution
Let n be a positive integer and 0 $\leq$ p $\leq$ 1. A random variable X has the binomial distribution with parameters n and p if the possible values
of X are {0, 1, . . . ,n} and the probabilities are


P( X = k ) =  $\binom{N}{k}$ $p^k$ $(1-p)^{n-k} $ for k = 0, 1, . . . ,n.


In the example below
- X = # of Heads from flipping a coin n times
- An outcome of of 0 = Tails
- An outcome of 1 = Heads
- P(Heads) = 0.65
- P(Tails) = 1 - 0.65 = 0.35

```r
p <- 0.65

a <- rbinom(50, 50, p)
hist(a, main="50 Coin Flips (p=0.65)", xlab = "Number of Heads", xlim=c(0, 50))

b <- rbinom(50, 50, p + 0.2) # p = 0.85
hist(b, main="50 Coin Flips (p=0.85)", xlab = "Number of Heads", xlim=c(0, 50))

c <- rbinom(100, 100, p - 0.3) # p = 0.35
hist(c, main="100 Coin Flips (p=0.35)", xlab = "Number of Heads", xlim=c(0, 100))

# The first two histograms below will reflect that more than half of our flips will be Heads.
# This makes sense since the coin has a bias that leans towards heads.
```

**Comment: I got confused with the definition of Bernoulli and Binomial since they sound very similar. **

Looking at the histograms in the examples above, the Bernoulli distribution reflects what we expect already. If I flip a coin 100 times and the probability of getting heads is 0.65, then I should get 65 heads. 

The Binomial distribution reflects the probability of getting an exact number of successes, not the success rate of Heads/Tails which the Bernoulli distribution shows. If the success rate of heads is 0.65, then the binomial distribution should show that if I flip a coin 100 times, I will be in the range of getting 65 heads (since 64, 66, 63, 67 heads are likely possible too).

### Geometric
Assume we're dealing with an infinite amount of trials. Let 0 $\leq$ p $\leq$ 1. A random variable X has the geometric distribution with success parameter p if the possible values of X are {1, 2, 3, . . . } and X satisfies 

P(X = k) = $(1-p)^{k - 1}$ p for positive integers k.

In the example below
- X = # of flips to get first Heads
- An outcome of of 0 = Tails
- An outcome of 1 = Heads
- P(Heads) = 0.65
- P(Tails) = 1 - 0.65 = 0.35

```r
a <- rgeom(100, 0.95)
hist(a, main="100 Experiments (p=0.95)", xlab = "Flips to get first Heads", ylim=c(0, 100))
# The first histogram reflects that it largely takes 0 flips to get the first head (meaning the first flip is Heads).
# Makes sense given that the prob of Heads is so high

b <- rgeom(100, 0.5)
hist(b, main="100 Experiments (p=0.5)", xlab = "Flips to get first Heads", ylim=c(0, 100))

c <- rgeom(100, 0.05)
hist(c, main="100 Experiments (p=0.05)", xlab = "Flips to get first Heads", ylim=c(0, 100))

# When comparing histogram b and c (prob of 0.50 vs 0.05), its easier to see that a low chance of success (0.05) will have a wider range
# of necessary flips until the first success
```

### Hypergeometric

Can't use a coin example anymore since the concept of "replacement" won't make sense. Instead our example below deals with a deck of cards that has black and red cards. Unusually, our deck of 52 cards has 36 red cards and 16 black cards.

```r
numRedCards <- 36
numBlackCards <- 16

a <- rhyper(100, numRedCards, numBlackCards, 20)
hist(a, main="100 Experiments (20 Draws)", xlab = "# of Red Cards")

b <- rhyper(100, numRedCards, numBlackCards, 30)
hist(b, main="100 Experiments (30 Draws)", xlab = "# of Red Cards")

c <- rhyper(100, numRedCards, numBlackCards, 40)
hist(c, main="100 Experiments (40 Draws)", xlab = "# of Red Cards")
```
