I'm not sure what the best way to visualize this is. I'm letting $ \ell = 100 $ and choosing to generate random real numbers from 0 to 100 and graph the result.



```python
# https://docs.scipy.org/doc/numpy-1.14.0/reference/routines.random.html - Documentation for random library + random.uniform

import matplotlib.pyplot as plt
import numpy as np

length = 100
z = np.random.uniform(0, length, 1000) # Z represents the location where the stick was broken
x = [min(breakpoint, length - breakpoint) for breakpoint in z] # x represents the length of the smaller stick

# Frequency Histogram
plt.hist(x, bins=100, range=(0, 100))
plt.title('Frequency of stick lengths')
plt.ylabel('Frequency of sticks of size = X')
plt.xlabel('X = Smaller Stick Length')
plt.show()

# Cumulative Frequency
plt.hist(x, bins=100, range=(0, 100), cumulative=True)
plt.title('Cumulative frequency of stick lengths')
plt.ylabel('Frequency of sticks of size <= X')
plt.xlabel('X = Smaller Stick Length')
plt.show()
```

- Our answers from part A concluded that the chance of finding a stick length greater than half the length (in this case 50) is always 1 (frequency = 1000).
- We also derived a formula that finding a smaller stick length, x, that is less than half is $ \frac{2x}{\ell} = \frac{x}{50} $, since $\ell = 100 $.

These statements make sense given that anything past length 50 shows a frequency of 1000. The second statement makes sense since increasing x, seems to have a proprtional (probably linear) increase in the frequency.
